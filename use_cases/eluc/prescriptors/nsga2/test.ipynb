```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd

from prescriptors.nsga2.torch_prescriptor import Candidate
from data import constants
from data.eluc_data import ELUCData
from data.torch_data import TorchDataset
```

```python
path = Path("prescriptors/nsga2/trained_prescriptors/test")
fig = plt.figure()
fig.patch.set_facecolor('white')
for gen in [1, 5, 10, 25, 50]:
    results_df = pd.read_csv(path / f"{gen}.csv")
    results_df = results_df[results_df["NSGA-II_rank"] == 1]
    results_df.sort_values(by="change", inplace=True)
    plt.plot(results_df["change"], results_df["ELUC"], label=f"gen {gen}", marker='o')
plt.legend()
plt.grid()
plt.show()
```
